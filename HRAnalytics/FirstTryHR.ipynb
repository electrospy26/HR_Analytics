## Evaluation

```python
def evaluation(string , model, ind_var, act):
    print(string.center(50,'*'))
    
    pred = model.predict(ind_var)
    cnf_matrix = confusion_matrix(act,pred)
    print('confusion matrix ',cnf_matrix)
    
    accuracy = accuracy_score(act,pred)
    print('Accuracy : ', accuracy)

    clf_report = classification_report(act,pred)
    print('Classification report:\n',clf_report)
    
evaluation('Test Data Evaluation ', rf_cl, x_test, y_test)
print('#'*50)
evaluation('Train Data Evauation ', rf_cl, x_train, y_train)
```

### KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn_cl = KNeighborsClassifier(n_neighbors=7, p=1)
knn_cl.fit(x_train,y_train)
```

```python
def evaluation(string,model, ind_var,y_act):
    print(string.center(50, '*'))
    
    pred = model.predict(ind_var)
    
    cnf_matrix = confusion_matrix(y_act, pred)
    print('Confusion Matrix :\n', cnf_matrix)
    
    print('**'*25)
    acc_score = accuracy_score(y_act, pred)
    print('Accuracy is :', acc_score)
    print('**'*25)
    
    clf_report = classification_report(y_act, pred)
    print('Classification Report :\n', clf_report)
    
evaluation('Test Data Evaluation', knn_cl, x_test, y_test)
```
