```python
df1=df
def catconsep(df):
    cat, con = [], []
    for i in df1.columns:
        if df1[i].dtype=='object':
            cat.append(i)
        else:
            con.append(i)
    return cat, con
```

```python
cat,con = catconsep(df)
print("Categorical features:\n",cat)
print('**'*50)
print('Continuous feature:\n',con)
```

```python
plt.figure(figsize=(20,20))
x = 1
plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.5, wspace=0.2)

for i in con:
    plt.subplot(9,3,x)
    x = x + 1
    sns.violinplot(data=df,x=i,y='Yes')
plt.show()
```
